## Steps 

to generate s channel SVJ signals, the workflow are

theory model -> madgraph -> Pythia8 -> Delphes -> Analysis codes 

We use madgraph to generate simulated process ->  output: LHE files  
pythia8 simulates the parton shower -> output: hepmc files   
delphes sumulates the respond of the detector -> output: root files   

Finally we use some analysis codes to extract information from root files and preprocess it. We'll get the jet images.

## Installation of software


#### Madgraph
You can simply download the package and unzip it. Then ./bin/mg
https://launchpad.net/mg5amcnlo/+download


#### Pythia
inside mg5_aMC environment, just type 
```  
MG5_aMC> install pythia8
```
MG5 will take care the configurations/path for you.

or you could install it manually and learn the standard way, for example 
```
./configure 
make -j8

```

#### Delphes

inside mg5_aMC environment, type 
```  
MG5_aMC> install Delphes
```


## Sample generation

``` 
./bin/mg5_aMC # you'll load the mg5_aMC 

MG5_aMC> import <PATH_TO_DMsimp_s_spin1>
MG5_aMC> define j = g u c t d b s g u~ c~ t~ d~ b~ s~
MG5_aMC> generate p p > xd xd~
MG5_aMC> add process p p > xd xd~ j
MG5_aMC> add process p p > xd xd~ j j
MG5_aMC> output sig_schannel
MG5_aMC> launch sig_schannel

```
Be sure you turn on Delphes and pythia8. 

![programs](./programs.png)

Then you can edit cards that contains many configurations:
![cards](./cards.png)


The easiest way is to replace the defalut cards( path: `./sig_schannel/Cards/` ) with those you download from:
https://github.com/hrzhao76/SVJ-madgraph/tree/master/data/sig_schannel

these four cards are important and you can only substitute these:  

>param_card.dat  
run_card.dat  
pythia8_card.dat  
delphes_card.dat  

You can also compare these cards with default ones and explore differences. 


Another important thing to do is to change Particle Date Group(PDG) IDs of the dark particles in the LHE files for `PYTHIA` to be able to recognize and shower these properly.

You'll find a `madevent_interface.py` from your downloaded files under the path `/sig_schannel/bin`, use it to replace  the defalut python script under the same path you generated. 

What this do is to add a snippet of codes to swap PDG IDs by: 
```
sed -i 's/5000521/4900101/g' <LHE filename>
``` 

Then after setting cards and handling the PDG ID changes, you can start generating samples.  
You'll get something like this: 
![results](./results.png)

This will create a root file `schannel_delphes_events.root` finally.

You are welcome to explore the Feynman diagrams and cross sections of different processes.  

Feynman diagrams: Go to Main Page and then click `Process Information`
